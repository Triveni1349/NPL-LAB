```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')

sentence =" the lion roars loudly in the forest at night always."

token=word_tokenize(sentence)
print("Tokens: ",token)
stop_words=set(stopwords.words("english"))
filtered_tokens=[word for word in token if word.lower() not in stop_words]
print("Filtered Token: ",filtered_tokens)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
nltk.download('wordnet')
sentence="The lion roars loudly in the forest at night always."

token=word_tokenize(sentence)
print(token)
stemmer=PorterStemmer()
Stemmer_tokens=[stemmer.stem(word) for word in token]
print(Stemmer_tokens)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

sentence =" the lion roars loudly in the forest at night always."

token=word_tokenize(sentence)
print("Tokens: ",token)
Lemmatizer=WordNetLemmatizer()
L_Tokens=[Lemmatizer.lemmatize(word) for word in token]
print("Lemmaword is :",L_Tokens)
```
